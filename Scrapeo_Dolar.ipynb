```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Genero una lista con las URL
urls = [
   "https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2024",
   "https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2024",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2024",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2023",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2022",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2021",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2020",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2019",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2018",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2017",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2016",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2015",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2014",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2013",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2012",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/diciembre-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/noviembre-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/octubre-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/septiembre-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/agosto-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/julio-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/junio-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/mayo-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/abril-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/marzo-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/febrero-2011",
"https://dolarhistorico.com/cotizacion-dolar-oficial/mes/enero-2011"



]



# Listas vacías para almacenar los datos de todas las URL
all_data = []

for url in urls:
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encuentro la tabla
        tabla_dolar = soup.find(class_="table table-bordered table-striped")

        # Extraigo los datos
        filas = tabla_dolar.find_all("tr")

        # Creo listas para cada columna
        Fecha = []
        compra = []
        venta = []
        variacion = []

        for fila in filas[1:]:  # Empiezo desde la segunda fila
            columnas = fila.find_all("td")

            # Verifico si hay elementos
            if columnas:
                Fecha.append(columnas[0].text.strip())
                compra.append(columnas[1].text.strip())
                venta.append(columnas[2].text.strip())
                variacion.append(columnas[3].text.strip())

        # Creo un DataFrame para cada URL y almaceno en la lista all_data
        df = pd.DataFrame({
            "Fecha": Fecha,
            "compra": compra,
            "venta": venta,
            "variacion": variacion})

        all_data.append(df)

# Concateno todos los DataFrames en uno solo
final_df_1 = pd.concat(all_data, ignore_index=True)


```

# Agrego Columnas con Informacion

```python
# Defino una función para convertir la fecha al formato "yyyy-mm-dd"
def convertir_fecha(fecha):
    return pd.to_datetime(fecha, format='%d/%m/%Y').strftime('%Y-%m-%d')

# Aplico la función a la columna "Fecha" y asigno el resultado a una nueva columna "Fecha_convertida"
final_df_1['Fecha_convertida'] = final_df_1['Fecha'].apply(lambda x: convertir_fecha(x))

# Agrego mas info al dataset
final_df_1.loc[(final_df_1["Fecha_convertida"] <= "2015-12-09"),"Presidente"] = "Cristina Fernández de Kirchner"
final_df_1.loc[(final_df_1["Fecha_convertida"] <= "2015-12-09"),"Partido Politico"] = "Frente para la Victoria"
final_df_1.loc[(final_df_1["Fecha_convertida"] >= "2015-12-10") & (final_df_1["Fecha_convertida"] <= "2019-12-09"), "Presidente"] = "Mauricio Macri"
final_df_1.loc[(final_df_1["Fecha_convertida"] >= "2015-12-10") & (final_df_1["Fecha_convertida"] <= "2019-12-09"), "Partido Politico"] = "Cambiemos"
final_df_1.loc[(final_df_1["Fecha_convertida"] >= "2019-12-10") & (final_df_1["Fecha_convertida"] <= "2023-12-09"), "Presidente"] = "Alberto Fernandez"
final_df_1.loc[(final_df_1["Fecha_convertida"] >= "2019-12-10") & (final_df_1["Fecha_convertida"] <= "2023-12-09"), "Partido Politico"] = "Unidad Ciudadana"
final_df_1.loc[(final_df_1["Fecha_convertida"] >= "2023-12-10"), "Presidente"] = "Javier Milei"
final_df_1.loc[(final_df_1["Fecha_convertida"] >= "2023-12-10"), "Partido Politico"] = "La Libertad Avanza"
```
